```python
import pandas as pd
import plotly.graph_objects as go

# Assuming 'volunteer_df' has a 'gender' column alongside 'Status'
# Filter and count for each gender separately
male_counts = volunteer_df[volunteer_df['gender'].str.lower() == 'male']['Status'].value_counts().reset_index()
female_counts = volunteer_df[volunteer_df['gender'].str.lower() == 'female']['Status'].value_counts().reset_index()

# Rename columns correctly and ensure 'Status' is treated as string
male_counts.columns = ['Status', 'Male']
female_counts.columns = ['Status', 'Female']
male_counts['Status'] = male_counts['Status'].astype(str)
female_counts['Status'] = female_counts['Status'].astype(str)

# Create a DataFrame of all unique statuses in the original data and convert to string
all_statuses = pd.DataFrame(volunteer_df['Status'].astype(str).unique(), columns=['Status'])

# Merge the counts into a single DataFrame on Status
status_counts = pd.merge(all_statuses, male_counts, on='Status', how='outer')
status_counts = pd.merge(status_counts, female_counts, on='Status', how='outer').fillna(0)

# Create a horizontal bar chart
fig = go.Figure(data=[
    go.Bar(name='Male', y=status_counts['Status'], x=status_counts['Male'], orientation='h', marker_color='#0057b7'),
    go.Bar(name='Female', y=status_counts['Status'], x=status_counts['Female'], orientation='h', marker_color='#ffd700')
])

# Change the bar mode to group to make the bars clustered
fig.update_layout(
    barmode='group',
    title='Comparison of Status Counts by Gender',
    xaxis_title='Count',
    yaxis_title='Status',
    width=800,  # width in pixels
    height=1100,  # height in pixels
    font=dict(family="Arial, sans-serif", color='black', size=12, weight='bold'),  # Set font properties including bold weight
    margin=dict(l=20, r=20, t=100, b=20)  # Adjust the top margin to give more space for the title
)

############## Ukraine-style theming. ###################
#economist_red = '#ffd700'
background_color = '#59caf2'  # 'rgb(195, 215, 227)'

# Customize the plot with the styled theme
fig.update_layout(
    title={'text': "Comparing Volunteer Status by Gender", 'y':0.97, 'x':0.5, 'xanchor': 'center', 'yanchor': 'top'},
    title_font=dict(size=18, color=economist_red, family="Arial, sans-serif", weight='bold'),
    plot_bgcolor=background_color,
    paper_bgcolor=background_color,
    legend_title="",
    xaxis_tickangle=-45
)
##########################################################

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Function to categorize age
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 21:
        return 'Under 21'
    elif age < 55:
        return 'Under 55'
    else:
        return '55 and over'

# Preprocess data
volunteer_df['Age Group'] = volunteer_df['age'].apply(categorize_age)
volunteer_df['gender'] = volunteer_df['gender'].str.upper()  # Ensure gender is uppercase

# Group the data
grouped = volunteer_df.groupby(['Status', 'gender', 'Age Group']).size().reset_index(name='Count')

# Define colors for each gender
gender_colors = {
    'MALE': '#0057b7',
    'FEMALE': '#ffd700'
}

# Create subplots: 1 row, 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=("MALE", "FEMALE"), horizontal_spacing=0.15)

# Adding traces, one subplot for each gender
genders = ['MALE', 'FEMALE']
for col, gender in enumerate(genders, start=1):
    for status in grouped['Status'].unique():
        for age_group in ['Under 21', 'Under 55', '55 and over']:
            subset = grouped[(grouped['Status'] == status) & (grouped['gender'] == gender) & (grouped['Age Group'] == age_group)]
            fig.add_trace(
                go.Bar(
                    y=[status],
                    x=subset['Count'],
                    name=f'{age_group}',
                    orientation='h',
                    marker=dict(color=gender_colors[gender], line=dict(color='black', width=3)),
                ),
                row=1, col=col
            )

# Update layout for better visual organization
background_color = '#59caf2'
fig.update_layout(
    title='Comparison of Status Counts by Gender and Age Group',
    yaxis_title='Status',
    xaxis_title='Count',
    yaxis2_title='Status',
    xaxis2_title='Count',
    height=600,
    width=1200,
    barmode='stack',
    legend_title_text='Age Group',
    plot_bgcolor=background_color,
    paper_bgcolor=background_color
)

fig.show()
```
